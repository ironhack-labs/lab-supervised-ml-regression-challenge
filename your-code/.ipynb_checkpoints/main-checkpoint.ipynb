## 1. Find the Naive Baseline

Before we build any models, let's establish a naive baseline. This will help us understand how well our models perform compared to a simple approach. In regression problems, the naive baseline is often the mean of the target variable.

### Task:
- Calculate the mean of the target variable 'charges'
- Explain why it's important to establish a naive baseline


````python
# Calculate the mean of the target variable 'charges'
naive_baseline = encoded_insurance['charges'].mean()

# Display the naive baseline
print(f"Naive Baseline (Mean of Charges): {naive_baseline:.2f}")
````

### IMPORTANT TO ESTABLISH A NAIVE BASELINE
- The naive baseline serves as a simple reference point for model performance.
- It represents the expected outcome if we predicted the same value (mean) for all instances.
- If a model performs worse than the naive baseline, it indicates the model isnt capturing useful patterns in the data.
- A good model should significantly outperform the naive baseline.
- The naive baseline assumes no knowledge about the features, only the target variables distribution.
- It highlights how much value the predictive model adds by utilizing feature information.

## 2. Initial Modelling Without GridSearch or Pipeline

Let's build a simple linear regression model without any feature engineering, grid search, or pipeline. This will serve as our initial baseline for comparison.

### Task:
- Split the data into training and test sets
- Train a simple linear regression model
- Evaluate its performance using regression metrics
- Write it down as a markdown below so you can keep track. This is a scientific experiment


````python
seed = 42
tst_size = 0.2 # 80% training size data, 20% test size data

def regression_metrics(test, pred, message):
    mae = mean_absolute_error(test, pred)
    mse = mean_squared_error(test, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(test, pred)
    
    # Display results
    print(message)
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"R Score: {r2:.2f}")
````

````python
# Separate features and target variable
X = encoded_insurance.drop(columns=['charges'])  # Features
y = encoded_insurance['charges']  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=tst_size, random_state=seed)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate the naive baseline
naive_baseline = y.mean()
baseline_mae = mean_absolute_error(y_test, [naive_baseline] * len(y_test))

# Display results
print(f"Naive Baseline (Mean Charges): {naive_baseline:.2f}")
print(f"Baseline MAE: {baseline_mae:.2f}\n")

regression_metrics(y_test, y_pred, "Linear Regression Model Performance:")
````

Linear Regression Model Performance:
Mean Absolute Error (MAE): 4901.38

## 3. Feature Engineering

Now, let's brainstorm and create some new features to see if we can improve the model's performance.

### Questions:
1. Should we create an interaction feature between 'bmi' and 'children'? 
2. Should we create age groups to see if the model improves by categorizing age?
3. Should we create a high-risk indicator based on 'smoker' and 'bmi'?

- Remember nothing is set in stone, this is your experiment, your hypothesis. You may not need to, but its important to explore these questions

### Task:
- Create new features based on the questions above
- Explain the rationale behind each feature



### 1. Should we create an interaction feature between bmi and children?
    - Interaction Feature Rationale:
        - bmi (Body Mass Index) measures body fat based on height and weight, and having children may impact medical costs due to additional healthcare needs.
        - By creating an interaction feature, we hypothesize that the combined effect of high BMI and having children could better predict charges.
  
### 2. Should we create age groups to see if the model improves by categorizing age?
    - Age Groups Rationale:
        - Categorizing continuous variables like age into groups can reveal trends or patterns that are not apparent in raw numbers.
        - Example age groups: Young (1830), Middle-aged (3150), and Older (51+).
        - This feature could improve the model by capturing non-linear effects of age on charges.
  
### 3. Should we create a high-risk indicator based on smoker and bmi?
    - High-Risk Indicator Rationale:
        - Smokers with high BMI may represent a group with significantly higher health risks and insurance costs.
        - Combining smoker and bmi into a binary indicator (e.g., 1 for high-risk, 0 otherwise) might help the model identify these costly cases.

````python
# Feature 1: Interaction between bmi and children
encoded_insurance['bmi_children'] = encoded_insurance['bmi'] * encoded_insurance['children']

# Feature 2: Age groups
def categorize_age(age):
    if age <= 30:
        return 'young'
    elif age <= 50:
        return 'middle-aged'
    else:
        return 'older'

encoded_insurance['age_group'] = encoded_insurance['age'].apply(categorize_age)

# Convert age_group to dummy variables (one-hot encoding)
encoded_insurance = pd.get_dummies(encoded_insurance, columns=['age_group'], drop_first=True)

# Feature 3: High-risk indicator based on smoker and bmi
encoded_insurance['high_risk'] = ((encoded_insurance['smoker_yes'] == 1) & (encoded_insurance['bmi'] > 30)).astype(int)

# Display the first few rows of the dataset with new features
encoded_insurance.head()
````

## 5. Modelling with Pipeline and Grid Search

Now, let's see how using pipelines can simplify our workflow and prevent data leakage. We'll also use GridSearchCV to find the best hyperparameters.

### Task:
- Create a pipeline that includes scaling and linear regression
- Define a parameter grid for hyperparameter tuning
- Use GridSearchCV to find the best parameters and evaluate the model performance


````python
# Define the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardize features
    ('model', Ridge())             # Ridge regression as the model
])

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'model__alpha': [0.01, 0.1, 1, 10, 100]  # Regularization strength for Ridge
}

# Perform Grid Search with Cross-Validation
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error')
grid_search.fit(X_train, y_train)

# Get the best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred = best_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Display results
print("Best Parameters:", best_params)

regression_metrics(y_test, y_pred, "\nModel Performance with Best Parameters:")
````

## 6. Trying Another Model with Pipeline

Let's try using a Gradient Boosting Regressor to see if it performs better.

### Task:
- Create and use a pipeline for Gradient Boosting Regressor
- Define a parameter grid for grid search
- Use GridSearchCV to find the best parameters and evaluate the model


````python
# Define the pipeline
pipeline = Pipeline([
    ('model', GradientBoostingRegressor(random_state=42))  # Gradient Boosting Regressor
])

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'model__n_estimators': [50, 100, 200],  # Number of boosting stages
    'model__learning_rate': [0.01, 0.1, 0.2],  # Learning rate shrinks the contribution of each tree
    'model__max_depth': [3, 5, 7],  # Maximum depth of the individual regression estimators
}

# Perform Grid Search with Cross-Validation
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred = best_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Display results
print("Best Parameters:", best_params)

regression_metrics(y_test, y_pred, "\nGradient Boosting Model Performance with Best Parameters:")
````

## 7. GridSearch with Several Models

Finally, let's compare several models using GridSearchCV to find the best one.

### Task:
- Define multiple models and their parameter grids
- Use GridSearchCV to find the best model and parameters


````python
# Define models and their parameter grids
models = {
    'Ridge': {
        'pipeline': Pipeline([('scaler', StandardScaler()), ('model', Ridge())]),
        'params': {'model__alpha': [0.01, 0.1, 1, 10, 100]}
    },
    'GradientBoosting': {
        'pipeline': Pipeline([('model', GradientBoostingRegressor(random_state=seed))]),
        'params': {
            # 'model__n_estimators': [50, 100, 200],
            # 'model__learning_rate': [0.01, 0.1, 0.2],
            # 'model__max_depth': [3, 5, 7]

            'model__n_estimators': [50, 300, 500],  # Increased number of trees
            'model__learning_rate': [0.01, 0.05, 0.1],  # learning rates
            'model__max_depth': [3, 5, 7],  # range of tree depths
            'model__subsample': [0.8, 1.0]  # add subsampling to prevent overfitting       
        }
    },
    'RandomForest': {
        'pipeline': Pipeline([('model', RandomForestRegressor(random_state=42))]),
        'params': {
            'model__n_estimators': [50, 100, 200],
            'model__max_depth': [None, 10, 20],
            'model__min_samples_split': [2, 5, 10]
        }
    }
}

# Perform GridSearch for each model
best_models = {}
for name, model_info in models.items():
    print(f"Running GridSearch for {name}...")
    grid_search = GridSearchCV(
        estimator=model_info['pipeline'],
        param_grid=model_info['params'],
        cv=5,
        scoring='neg_mean_absolute_error',
        n_jobs=-1
    )
    grid_search.fit(X_train, y_train)
    best_models[name] = {
        'best_estimator': grid_search.best_estimator_,
        'best_params': grid_search.best_params_ if model_info['params'] else "Default",
        'best_score': -grid_search.best_score_  # Convert negative score back to positive
    }

# Evaluate the best model from each type
results = []
for name, model_info in best_models.items():
    best_model = model_info['best_estimator']
    y_pred = best_model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results.append({
        'Model': name,
        'Best Parameters': model_info['best_params'],
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2
    })

# Display the results
results_df = pd.DataFrame(results)
print(results_df)

# Select the best model based on MAE
best_overall_model = results_df.loc[results_df['MAE'].idxmin()]
print("\nBest Overall Model:")
print(best_overall_model)
````

````python
# Refines the hyperparameters 
refined_param_grid = {
    'model__n_estimators': [50, 300, 500],  # Increased number of trees
    'model__learning_rate': [0.01, 0.05, 0.1],  # learning rates
    'model__max_depth': [3, 5, 7],  # range of tree depths
    'model__subsample': [0.8, 1.0]  # add subsampling to prevent overfitting
}

# Creates a pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    ('model', GradientBoostingRegressor(random_state=seed))  
])

# Performs GridSearchCV
grid_search_refined = GridSearchCV(pipeline, 
                                   refined_param_grid, 
                                   cv=5, # Folds
                                   scoring='r2', 
                                   verbose=1)

grid_search_refined.fit(X_train_simplified, y_train)

# Displays the best parameters from refined tuning
print("Refined Best Parameters:", grid_search_refined.best_params_)

# Evaluates the refined model on the test set
refined_best_model = grid_search_refined.best_estimator_
y_pred_refined = refined_best_model.predict(X_test_simplified)

# Calculates the performance metrics for the refined model
mae_refined = mean_absolute_error(y_test, y_pred_refined)
mse_refined = mean_squared_error(y_test, y_pred_refined)
r2_refined = r2_score(y_test, y_pred_refined)
rmse_refined = np.sqrt(mean_squared_error(y_test, y_pred_refined))

print("Refined Test Set:")
print(f"  MAE: {mae_refined:.4f}")
print(f"  MSE: {mse_refined:.4f}")
print(f"  RMSE: {rmse_refined:.2f}")
print(f"  R2: {r2_refined:.2f}")
````

````python
# Adjust R2
def adjusted_r2(r2, n, p):
    """
    Calculate Adjusted R2
    Args: 
        r2 (float): r2 refined
        n (int): test data size
        p (int): number of features
    Return:
        adjusted r2 (float)
    """
    return 1 - ((1 - r2) * (n - 1) / (n - p - 1))


# Check the median absolute error and adjusted R2.
median_ae = median_absolute_error(y_test, y_pred_refined) # Median Absolute Error
print(f"Median Absolute Error (MedAE): {median_ae:.2f}")

# Calculates the adjusted R2
n = len(y_test)  # Number of observations
p = X_test_simplified.shape[1]  # Number of features
adj_r2 = adjusted_r2(r2_refined, n, p)
print(f"Adjusted R2: {adj_r2:.2f}")
````

### Explanation:
The RMSE provides an estimate of the average error in your model's predictions. We will also analyze the errors by categorizing them into underestimations and overestimations to understand their financial impact.

#### Steps to Calculate Underestimation and Overestimation Errors:

1. **Calculate RMSE**:
   - Use the `mean_squared_error` function from `sklearn.metrics` and pass your actual values (`y_test`) and predicted values (`y_pred_final`) to it.
   - Take the square root of the result to get the RMSE.
   
2. **Calculate Underestimation Error**:
   - Identify the instances where the actual charges (`y_test`) are greater than the predicted charges (`y_pred_final`).
   - For these instances, calculate the difference between the actual and predicted charges.
   - Sum these differences to get the total underestimation error.

3. **Calculate Overestimation Error**:
   - Identify the instances where the actual charges (`y_test`) are less than the predicted charges (`y_pred_final`).
   - For these instances, calculate the difference between the predicted and actual charges.
   - Sum these differences to get the total overestimation error.

4. **Calculate Total Potential Cost or Loss**:
   - Add the total underestimation error and the total overestimation error to get the total potential cost or loss.

#### Detailed Instructions:

1. **Calculate RMSE**:
   - Use `mean_squared_error` with `y_test` and `y_pred_final`.
   - Use `np.sqrt` to take the square root of the result.

2. **Calculate Underestimation Error**:

   - Subtract the predicted values from the actual values for these instances.
   - Sum these differences.

3. **Calculate Overestimation Error**:
   - Use a boolean condition to filter `y_test` values that are less than `y_pred_final`.
   - Subtract the actual values from the predicted values for these instances.
   - Sum these differences.

4. **Calculate Total Potential Cost or Loss**:
   - Add the results of the underestimation error and overestimation error to get the total potential cost or loss.

### Example Walkthrough:

   - `rmse = np.sqrt(mean_squared_error(y_test, y_pred_final))`
   - This gives you the average prediction error in dollars.

2. **Calculate Underestimation Error**:

   - This gives you the total amount by which the model undercharged.

   - This gives you the total amount by which the model overcharged.

4. **Calculate Total Potential Cost or Loss**:
   - `total_potential_loss = underestimation_error + overestimation_error`
   - This gives you the total financial impact of the model's errors.
